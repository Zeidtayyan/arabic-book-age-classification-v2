```python
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side

age_mapping = {
    # ٣-٨
    'الأرنب والسلحفاة': '٣-٨',
    'البطة-القبيحة': '٣-٨',
    'البطة القبيحة': '٣-٨',
    'الخنازير-الثلاثة': '٣-٨',
    'الخنازير الثلاثة': '٣-٨',
    'حورية تقول الصدق': '٣-٨',
    'الراعي الكذاب': '٣-٨',
    'القصة المصورة الراعي الكذاب': '٣-٨',
    'السلحفاة الطائرة': '٣-٨',
    'الفيلة الجميلة': '٣-٨',
    'الكلب الصامت': '٣-٨',
    'النسناس فرناس': '٣-٨',
    'بذرة التفاح': '٣-٨',
    'الجرة العجيبة': '٣-٨',
    'كنز الثعالب': '٣-٨',
    'كنز الثعلب': '٣-٨',
    'قصة الثور والضفدع المغور': '٣-٨',
    'الثور والضفدع': '٣-٨',

    # ٩-١٢
    'الأميرة الصغيرة(سالي)': '٩-١٢',
    'الأميرة الصغيرة': '٩-١٢',
    'بلا عائلة(ريمي)': '٩-١٢',
    'بلا عائلة': '٩-١٢',
    'الصبيّ_الأعرج': '٩-١٢',
    'الصبيّ الأعرج': '٩-١٢',
    'الصبي الأعرج': '٩-١٢',
    'الأغصان': '٩-١٢',
    'جزيرة الكنز': '٩-١٢',
    'جزيرة_الكنز': '٩-١٢',
    'هايدي سنوات التعلم والترحال': '٩-١٢',
    'مغامرات توم سوير': '٩-١٢',
    'روبنسون كروزو': '٩-١٢',
    'عائلة روبنسون السويسرية': '٩-١٢',
    'علي بابا و الأربعون لصا': '٩-١٢',
    'علي بابا والأربعون': '٩-١٢',
    'السندباد البحري': '٩-١٢',
    'كامل كيلاني-السندباد البحري': '٩-١٢',
    'arabicpdfs.com-عَلَاءُ-الدِّينِ': '٩-١٢',
    'علاء الدين': '٩-١٢',
    'حكاية التاجر مع العفريت': '٩-١٢',
    'أليس في بلاد العجائب': '٩-١٢',
    'الحديقة_السرية': '٩-١٢',
    'الحديقة السرية': '٩-١٢',
    'أبو الحسن': '٩-١٢',
    'أبو الحسن1': '٩-١٢',
    'شهرزاد بنت الوزير': '٩-١٢',
    'حول العالم في ثمانين يوماً': '٩-١٢',
    'قصص النبيين للأطفال': '٩-١٢',
    'مغامرات أنا بينوكيو': '٩-١٢',
    'مغامرات في أعماق البحار': '٩-١٢',
    'عقلة الإصبع': '٩-١٢',
    'عقلة الأصبع': '٩-١٢',

    # ١٣-١٨
    'فاتتني صلاة': '١٣-١٨',
    'في_سبيل_التاج': '١٣-١٨',
    'في سبيل التاج': '١٣-١٨',
    'الفضيلة': '١٣-١٨',
    'ماجدولين': '١٣-١٨',
    'الأيام': '١٣-١٨',
    'رواية المشكلة الأخيرة': '١٣-١٨',
    'المشكلة الأخيرة': '١٣-١٨',
    '-سيرة مدينة عمان في الأربعينيات': '١٣-١٨',
    'سيرة مدينة عمان': '١٣-١٨',
    'الأدب الصغير': '١٣-١٨',
    'رواية جريمة فى قطار الشرق': '١٣-١٨',
    'جريمة في قطار الشرق': '١٣-١٨',
    'رواية ثم لم يبق أحد': '١٣-١٨',
    'ثم لم يبقَ أحد': '١٣-١٨',
    'رواية مقتل روجر أكرويد': '١٣-١٨',
    'مقتل روجر أكرويد': '١٣-١٨',
    'مغامرة الجوهرة الزرقاء': '١٣-١٨',
    'مغامرة_الجوهرة_الزرقاء': '١٣-١٨',
    'مغامرة العصابة الرقطاء': '١٣-١٨',
    'مغامرة_تاج_الزمرد': '١٣-١٨',
    'مغامرة تاج الزمرد': '١٣-١٨',
    'نظرية الفستق': '١٣-١٨',
    'في قلبي أنثى عبرية': '١٣-١٨',
    'خبز الفداء': '١٣-١٨',
    'حي_بن_يقظان': '١٣-١٨',
    'حي بن يقظان': '١٣-١٨',
    'عرس الزين': '١٣-١٨',
    'العادات السبع للمراهقين الأكثر فعالية': '١٣-١٨',
    'جثة فى المكتبة': '١٣-١٨',
    'جثة بالمكتبة': '١٣-١٨',
    'فضيحة_في_بوهيميا': '١٣-١٨',
    'فضيحة_في_بوهيميا (1)': '١٣-١٨',
    'فضيحة في بوهيميا': '١٣-١٨',
    'صاحب الظل الطويل': '١٣-١٨',
    'خمسون شمعة لإضاءة دروبكم': '١٣-١٨',

    # ١٨-٢١
    'المعذبون في الأرض': '١٨-٢١',
    'كان ما كان ميخائيل نعيمة': '١٨-٢١',
    'العاقر': '١٨-٢١',
    'موسم الهجرة إلى الشمال': '١٨-٢١',
    'الكرنك': '١٨-٢١',
    'مدن الملح': '١٨-٢١',
    'شرق المتوسط': '١٨-٢١',
    'اللص والكلاب': '١٨-٢١',
    'دعاء الكروان': '١٨-٢١',
    'بداية ونهاية': '١٨-٢١',
    'قصة حب مجوسية': '١٨-٢١',
    'مجوسية': '١٨-٢١',
    'أرض السواد': '١٨-٢١',
    'الاشجار واغتيال مرزوق': '١٨-٢١',
    'الأدب_الكبير': '١٨-٢١',
    'الأدب الكبير': '١٨-٢١',
    'الحيوان': '١٨-٢١',
    'الحب الضائع': '١٨-٢١',
    'قصر الشوق': '١٨-٢١',
    'بين القصرين': '١٨-٢١',
    'السكرية': '١٨-٢١',
    'كتاب البخلاء للجاحظ': '١٨-٢١',
    'البخلاء': '١٨-٢١',
    'القلق كتاب': '١٨-٢١',
    'بنك المعلومات': '١٨-٢١',
    'بنك القلق': '١٨-٢١',
    'القاهرة الجديدة': '١٨-٢١',
    'زقاق_المدق': '١٨-٢١',
    'زقاق المدق': '١٨-٢١',
    'خان الخليلي': '١٨-٢١',
    'الأخوة كارامازوف': '١٨-٢١',
    '2 الجريمة والعقاب': '١٨-٢١',
    'العقاب': '١٨-٢١',
    'الجريمة والعقاب': '١٨-٢١',

    # ٢١+
    'هكذا تكلم زرادشت': '٢١+',
    'فصوص الحكم أبن عربي': '٢١+',
    'فصوص الحكم': '٢١+',
    'رسالة_الغفران': '٢١+',
    'رسالة الغفران': '٢١+',
    'أسطورة سيزيف': '٢١+',
    'امانويل كانط نقد العقل المحض': '٢١+',
    'نقد العقل المحض': '٢١+',
    'العالم إرادة وتمثلا': '٢١+',
    'النار_في_التحليل_النفسي': '٢١+',
    'التحليل النفسي للنار': '٢١+',
    'المواقف_والمخاطبات': '٢١+',
    'المواقف والمخاطبات': '٢١+',
    'باروخ سبينوزا-علم الأخلاق': '٢١+',
    'باروخ سبينوزا': '٢١+',
    'الأخلاق': '٢١+',
    'تهافت التهافت': '٢١+',
    'عبد القاهر الجرجاني-دلائل الإعجاز': '٢١+',
    'دلائل الإعجاز': '٢١+',
    'التطور المبدع هنري برغسون': '٢١+',
    'التطور المبدع': '٢١+',
    'الإمتاع والمؤانسة': '٢١+',
    'الشفاء (الإلهیات) - إبن سينا': '٢١+',
    'الشفاء': '٢١+',
    'رسائل إخوان الصفا وخلان الوفا': '٢١+',
    'العقد الإجتماعي - جان جاك روسو': '٢١+',
    'العقد الاجتماعي': '٢١+',
    'إرادة القوة - نيتشه': '٢١+',
    'إرادة القوة': '٢١+',
    'ابن سلام الجمحي-طبقات فحول الشعراء': '٢١+',
    'طبقات فحول الشعراء': '٢١+',
    'الأمواج': '٢١+',
    'الخصائص': '٢١+',
    'الصخب والعنف': '٢١+',
    'أنّا كارنينا': '٢١+',
    'آنا كارنينا': '٢١+',
    'البحث عن الزمن المفقود': '٢١+',
    'بحثاً عن الزمن الضائع': '٢١+',
    'السيدة دالاوي': '٢١+',
    'من الفناء إلى البقاء': '٢١+',
}

# Color map for age categories
age_colors = {
    '٣-٨': 'C6EFCE',      # Light green
    '٩-١٢': 'BDD7EE',     # Light blue
    '١٣-١٨': 'FCE4D6',    # Light orange
    '١٨-٢١': 'E2EFDA',    # Light sage
    '٢١+': 'D9D2E9',      # Light purple
}

# ===== تعديل المسار حسب مكان الملف عندك =====
FILE_PATH = '/content/drive/MyDrive/zaid t/all_books.xlsx'
OUTPUT_PATH = '/content/drive/MyDrive/zaid t/all_books_W_L.xlsx'

# Load workbook
wb = load_workbook(FILE_PATH)
ws = wb.active

# Add header for age category
ws.cell(row=1, column=3, value='Age Category')
ws.cell(row=1, column=3).font = Font(bold=True, size=12, name='Arial', color='FFFFFF')
ws.cell(row=1, column=3).fill = PatternFill('solid', fgColor='4472C4')
ws.cell(row=1, column=3).alignment = Alignment(horizontal='center', vertical='center')

# Style existing headers
for col in [1, 2]:
    cell = ws.cell(row=1, column=col)
    cell.font = Font(bold=True, size=12, name='Arial', color='FFFFFF')
    cell.fill = PatternFill('solid', fgColor='4472C4')
    cell.alignment = Alignment(horizontal='center', vertical='center')

# Match books with age categories
matched = 0
unmatched_books = set()

for row in range(2, ws.max_row + 1):
    book_name = str(ws.cell(row=row, column=1).value or '')
    age = None

    # Try exact match first
    if book_name in age_mapping:
        age = age_mapping[book_name]
    else:
        # Try partial/contains match
        for key, val in age_mapping.items():
            if key in book_name or book_name in key:
                age = val
                break

    if age:
        cell = ws.cell(row=row, column=3, value=age)
        cell.font = Font(size=11, name='Arial', bold=True)
        cell.alignment = Alignment(horizontal='center', vertical='center')
        cell.fill = PatternFill('solid', fgColor=age_colors.get(age, 'FFFFFF'))
        matched += 1
    else:
        unmatched_books.add(book_name)

# Set column width
ws.column_dimensions['C'].width = 15

# Add thin border to all data cells
thin_border = Border(
    left=Side(style='thin'),
    right=Side(style='thin'),
    top=Side(style='thin'),
    bottom=Side(style='thin')
)
for row in range(1, ws.max_row + 1):
    for col in range(1, 4):
        ws.cell(row=row, column=col).border = thin_border

wb.save(OUTPUT_PATH)

print(f"✅ Matched rows: {matched}")
print(f"📊 Total rows: {ws.max_row - 1}")

if unmatched_books:
    print(f"\n⚠️ Unmatched books ({len(unmatched_books)}):")
    for b in sorted(unmatched_books):
        print(f"  - '{b}'")
else:
    print("\n🎉 All books matched successfully!")
```
